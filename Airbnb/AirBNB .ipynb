```python
import pandas as  pd
import numpy as np
```

```python
listing= pd.read_csv(r'D:/python projects/Airbnb+Data/Airbnb Data/Listings.csv', encoding='ISO-8859-1',low_memory=False,parse_dates=["host_since"])
listing.head()
```

```python
listing.info()
```

```python
paris_listing= (listing.query("city== 'Paris'").loc[:,["host_since","neighbourhood","city","accommodates","price"]])
paris_listing.info()
```

```python
paris_listing.isna().sum()
```

```python
paris_listing.describe()
```

```python
paris_listing.query("accommodates==0 and price==0").count()
```

```python
paris_listing_neghbourhood=(paris_listing.groupby('neighbourhood').agg({"price":"mean"}).sort_values("price"))

paris_listing_neghbourhood
```

```python
paris_listing_accomdates=  ( paris_listing.query("neighbourhood=='Elysee'")
                            .groupby("accommodates")
                            .agg({"price": "mean"}).sort_values("price")
                           
                           )

paris_listing_accomdates.head()
```

```python
paris_listing_over_time=(
      paris_listing.set_index("host_since").resample("Y")
    .agg({"neighbourhood":"count","price":"mean"
         })
)
paris_listing_over_time
```

```python
import seaborn as sns 
(paris_listing_neghbourhood.plot.barh( title="Avg listing price by Paris Neighbourhood",
                                     xlabel="Price oer Night(Euros)",
                                      ylabel="Neighbourhood",
                                      legend=None
                                     )
)
sns.despine()
```

```python
import seaborn as sns 
(paris_listing_accomdates.plot.barh( title="Avg listing price by Paris Accomodate Number",
                                     xlabel="Price per Night(Euros)",
                                      ylabel="Accommodation capacity",
                                      legend=None
                                     )
)
sns.despine()
```

```python
paris_listing_over_time["neighbourhood"].plot(
    ylabel="New hosts",
    title="New AirBNB Hosts in Paris over time"
    
)
sns.despine()
```

```python
paris_listing_over_time["price"].plot(
    ylabel="AVG price",
    title="Aereage price AirBNB Hosts in Paris over time"
    
)
sns.despine()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(
    paris_listing_over_time.index,paris_listing_over_time["neighbourhood"],label ="New Hosts",
    c="Blue"
)
ax.set_ylabel("New Hosts")
ax2=ax.twinx()
ax2.plot(paris_listing_over_time.index,paris_listing_over_time["price"],label="Price",c="Pink")
ax2.set_ylabel("Average Price")
ax2.set_ylim(0)
ax.set_title("2015 Regualtions lead to lower New Hosts , Higher Prices")
```
